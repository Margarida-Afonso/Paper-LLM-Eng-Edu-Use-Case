### 4️⃣ Function with List Parameter
Write a function `sum_list(lst)` that receives a list of numbers and returns the sum of all elements.
Test it with the list `[4, 7, 2, 9]`.


````python
def sum_list(lst):
    total = 0
    for num in lst:
        total += num
    return total

print(sum_list([4, 7, 2, 9]))  # Output: 22
````
